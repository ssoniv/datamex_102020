```python
import random

class Soldier:
    def __init__(self):
        self.health = 0
        self.power = 0
        self.specialnum = 0
        
    def receiveDamage(self, damage):
            self.health -= damage   

class mage(Soldier):
    def __init__(self):
        Soldier.__init__(self)
        self.job = 'Mage'
        self.health = 200
        self.power = 50
        self.agility = 2
               
    def special(self):
        if self.specialnum==0:
            self.power *=2
            self.health*=.75
            self.specialnum+=1
        else:
            print('Cant use special')
        
    def end_turn(self):
        self.power = 50
        self.agility = 2   
        
class archer(Soldier):
    def __init__(self):
        Soldier.__init__(self)
        self.job = 'Archer'
        self.health = 250
        self.power = 20
        self.agility = 4
        
        
    def special(self):
        if self.specialnum==0:
            self.agility *=2
            self.power*=.75
            self.specialnum+=1
        else:
            print('Cant use special')
           
    def end_turn(self):
        self.power = 20
        self.agility = 4      

class swordsman(Soldier):
    def __init__(self):
        Soldier.__init__(self)
        self.job = 'swordsman'
        self.health = 300
        self.power = 30
        self.agility = 2
        
        
    def special(self):
        if self.specialnum==0:
            self.health*=1.25
            self.power*=.75
            self.specialnum+=1
        else:
            print('Cant use special. lost turn')
     
    def end_turn(self):
        self.power = 30
        self.agility = 2.5
        
        
        
player_1= mage()
player_2=archer()
        
    
    
while player_1 != "mage" or player_1 != "swordsman" or player_1 !="archer":
    player_1=input("CHOOSE YOUR WARRIOR!!! \n\nType either 'mage' , 'swordsman' or 'archer': ").lower()
    if player_1 == "mage":
        print()
        print("You are a Mage.")
        print("Your special increases your power but decreases your health")
        print()
        player_1 =mage()
    elif player_1 == "swordsman":
        print()
        print(f"You are a Swordsman.")
        print("Your special increases your health but decreases your power")
        print()
        player_1 = swordsman()
    elif player_1 == "archer":
        print()
        print(f"You are an Archer.")
        print("Your special increases your agility but decreases your power")
        print()
        player_1 = archer()
    else:
        print("\nInvalid selection. Please type either 'mage' , 'swordsman' or 'archer':.\n")
        print("\n-----------------------------------------\n")
        continue
    break    
    
    
player_2 = random.choice([mage(), archer(), swordsman()])    
print(f'YOU WILL BE FIGHTING AGAINST ONE {player_2.job}')

        
###TURN####
contador_1=0
contador_2=0
defense_1=1
defense_2=1

while player_1.health>0 and player_2.health>0:
    contador_1 += player_1.agility
    contador_2 += player_2.agility
    if contador_1>=10 and player_1.health>0:
        contador_1-= 10
        print('-------------YOUR TURN----------------')
        choice_1=input('Choose attack, special or defend: ')
        if choice_1== 'defend':
            defense_1=.5
            print("YOU defend")
        elif choice_1 == 'attack':
            print("YOU attack")
            player_2.receiveDamage(player_1.power * defense_2)
            print(f'Opponent receives {player_1.power*defense_2} damage.')
            defense_2 = 1
            if player_2.health<=0:
                print("Opponent is dead")
                break 
        elif choice_1 =='special':
            player_1.special()
            print("YOU use your special, your stats go up")
        else:
            "failed"
    elif contador_2>=10 and player_2.health>0:
        contador_2-= 10
        print('-----------OPPONENTs TURN-------------')
        choice_2 = random.choice(['attack', 'defend', 'special']) 
        if choice_2== 'defend':
            defense_2=.5
            print("Opponent defends")
            
        elif choice_2 == 'attack':
            print("Opponent attacks")
            player_1.receiveDamage(player_2.power * defense_2)
            print(f'YOU receive {player_2.power*defense_1} damage.')
            defense_1 = 1
            if player_1.health<=0:
                print("YOU ARE dead")
                
                break 
        elif choice_2 == 'special':
            player_2.special()
            print("Opponent used special, stats up")
        else:
            "failed"
                
    if player_1.specialnum>0:
        player_1.specialnum+= 1
        if player_1.specialnum==4:
            player_1.specialnum=0
            print('++++YOUR STATS GO BACK TO NORMAL!!!')
            player_1.end_turn()
    if player_2.specialnum>0:
        player_2.specialnum+= 1
        if player_2.specialnum==4:
            player_2.specialnum=0
            print('++++OPPONENTS STATS GO BACK TO NORMAL!!!')
            player_2.end_turn()

   









```
