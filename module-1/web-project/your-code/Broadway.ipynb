```python
def get_original_cast(url):
    req = requests.get(url)
    
    soup = BeautifulSoup(req.content,'html.parser')
    cast = soup.find(id='OpeningNightCast')
    p = re.compile('(?<=\\n)\w+\s\w*(?=\\n)')
    opening_cast = p.findall(cast.text)
    
    return opening_cast 
```

```python
get_original_cast(url)[0:5]
```

```python
###The main problem is that we don't know how the id number 
#of each show is given so we need a list of the links

url = 'https://www.ibdb.com/Grosses/ViewProduction'
req = requests.get(url)
soup = BeautifulSoup(req.content,'html.parser')
###There is a bigger list inside the database but we need to do some 
### searches 
##for the time being we will use this ones
```

```python
##find all urls on the main page
lst=[]
for link in soup.find_all('a'):
    if link.get('href').startswith('/broadway'):
        lst.append(link.get('href'))
```

```python
#find all the name of the shows
for i in range(len(lst)):
    lst[i]=[soup.find_all('i')[18+i].text,lst[i]]
lst= lst[0:36]
```

```python
#A function that checks all the urls
def url_checker(url_list):
    url= 'https://www.ibdb.com'+url_list[1]
    cast= get_original_cast(url)
    unwanted = {'Broadway debut'}
    cast = [e for e in cast if e not in unwanted]
    return cast[0:5]
    
```

```python
df = pd.DataFrame() 
for i in range(len(lst)):
    try:
        df[lst[i][0]]=url_checker(lst[i])
    except:
        pass
    
df
##we have our list of actors 
```

```python
df2 = df.T
df2.index.name ="Show_Name"
```

```python
gross_actors = pd.merge(df1, df2, on = 'Show_Name')
gross_actors.sample(10)
```
