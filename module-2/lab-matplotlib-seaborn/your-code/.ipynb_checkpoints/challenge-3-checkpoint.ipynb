#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
fig,axes = plt.subplots(1,3,figsize=(10,5))
axes[0].hist(titanic.Age, bins=10, histtype='bar', rwidth=0.8)
axes[1].hist(titanic.Age, bins=20, histtype='bar', rwidth=0.8)
axes[2].hist(titanic.Age, bins=50, histtype='bar', rwidth=0.8);
```

#### Whether the bin size affects your plot? Why?

```python
#It does, the more bins we have to closer to the actual data. and we might see some ages missing or low
#but the bigger the sample the less noticeable differences
```
